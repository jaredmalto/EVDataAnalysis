```python
import pandas as pd
import matplotlib.pyplot as plt

#import csv as dataframe
df = pd.read_csv("Battery Test Data.csv")
```

```python
internal_resist_measure_1 = df["Internal Resistance Measure 1"]
```

```python
quality_parameter = df["Quality Parameter"]
```

```python
# use a scatter plot to find correlation
plt.scatter(internal_resist_measure_1, quality_parameter)
# rename axes
plt.xlabel("Internal Resistance Measure 1")
plt.ylabel("Quality Parameter")
#failure threshold
plt.axhline(y = -7.5)
```

```python
internal_resist_measure_2 = df["Internal Resistance Measure 2"]
```

```python
plt.scatter(internal_resist_measure_2, quality_parameter)
plt.xlabel("Internal Resistance Measure 2")
plt.ylabel("Quality Parameter")
plt.axhline(y = -7.5)
```
